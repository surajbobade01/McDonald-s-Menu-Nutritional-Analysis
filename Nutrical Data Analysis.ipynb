Exploratory Data Analysis (EDA)
Analyze the distribution of calorie counts across menu items.
Explore the nutritional content (e.g., fat, protein, carbohydrates) of different items.
Identify trends and patterns in the dataset.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], kde=True)
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.show() # Distribution of calories across all items
```

```python
plt.figure(figsize=(12, 6))
df.groupby('Category')['Calories'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Calories by Food Category')
plt.xlabel('Category')
plt.ylabel('Average Calories')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()  # Create a bar plot of average calories by category
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Total Fat', y='Calories')
plt.title('Calories vs Total Fat')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.show()  # Create a scatter plot of Calories vs Total Fat
```

```python
avg_nutrition = df.groupby('Category')[['Calories', 'Total Fat', 'Carbohydrates', 'Protein']].mean()
print("\
Average nutritional values by category:")
print(avg_nutrition)
```

```python
avg_nutrition_stacked = avg_nutrition[['Total Fat', 'Carbohydrates', 'Protein']]
avg_nutrition_stacked.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Average Macronutrient Composition by Category')
plt.xlabel('Category')
plt.ylabel('Grams')
plt.legend(title='Macronutrient', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()  #  Stacked bar chart of macronutrients by category
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.select_dtypes(include=[float, int]).corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of Nutritional Values')
plt.tight_layout()
plt.show()  #Correlation heatmap
```
