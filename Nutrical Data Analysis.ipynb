Exploratory Data Analysis (EDA)
Analyze the distribution of calorie counts across menu items.
Explore the nutritional content (e.g., fat, protein, carbohydrates) of different items.
Identify trends and patterns in the dataset.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Distribution of Calories Counts accross menu item')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
nutritional_columns = ['Total Fat', 'Protein', 'Carbohydrates']

plt.figure(figsize=(15, 6))
for i, column in enumerate(nutritional_columns):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(x='Category', y=column, data=df)
    plt.title(f'Box Plot Of {column} By Category')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(15,8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap= 'coolwarm', square=True)
plt.title('Correlation Matrix of Nutritional Components')
plt.show()  
```
